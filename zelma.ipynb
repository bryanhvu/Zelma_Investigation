## Data Filtering

```python
df_wi = df_wi[df_wi['DataLevel'] == 'School']
df_ky = df_ky[df_ky['DataLevel'] == 'School']
df_dc = df_dc[df_dc['DataLevel'] == 'School']
df_tx = df_tx[df_tx['DataLevel'] == 'District']
```

```python
subjects = ['math', 'ela']

df_wi = df_wi[df_wi['Subject'].isin(subjects)]
df_ky = df_ky[df_ky['Subject'].isin(subjects)]
df_dc = df_dc[df_dc['Subject'].isin(subjects)]
df_tx = df_tx[df_tx['Subject'].isin(subjects)]
```

```python
df_wi = df_wi[df_wi['StudentSubGroup_TotalTested'].ne(0)]
df_ky = df_ky[df_ky['StudentSubGroup_TotalTested'].ne(0)]
df_dc = df_dc[df_dc['StudentSubGroup_TotalTested'].ne(0)]
df_tx = df_tx[df_tx['StudentSubGroup_TotalTested'].ne(0)]
```
